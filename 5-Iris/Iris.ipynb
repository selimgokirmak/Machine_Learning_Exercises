```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
species = data.iloc[:,-1].values
species[:] = le.fit_transform(species[:])
```

```python
data["Species"] = pd.Series(species[:])
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data["Species"] = [int(i) for i in data["Species"]]
```

```python
data.info()
```

```python
data = data.iloc[:,1:]
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data["Species"].value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
sns.distplot(data["SepalLengthCm"])
plt.subplot(2,2,2)
sns.scatterplot(data=data, x=data.index, y=data["SepalLengthCm"], hue="Species")
plt.subplot(2,2,3)
sns.boxplot(data['SepalLengthCm'])
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
sns.distplot(data["SepalWidthCm"])
plt.subplot(2,2,2)
sns.scatterplot(data=data, x=data.index, y=data["SepalWidthCm"], hue="Species")
plt.subplot(2,2,3)
sns.boxplot(data['SepalWidthCm'])
```

```python
for column in data.columns[:-1]:
  for spec in data["Species"].unique():
    selected_spec = data[data["Species"] == spec]
    selected_column = selected_spec[column]
    
    std = selected_column.std()
    avg = selected_column.mean()
    
    tsp = avg + 3*std
    tsm = avg - 3*std
    
    outliers = selected_column[((selected_spec[column] > tsp) | (selected_spec[column] < tsm))].index
    
    data.drop(index=outliers, inplace=True)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
sns.distplot(data["PetalLengthCm"])
plt.subplot(2,2,2)
sns.scatterplot(data=data, x=data.index, y=data["PetalLengthCm"], hue="Species")
plt.subplot(2,2,3)
sns.boxplot(data['PetalLengthCm'])
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
sns.distplot(data["PetalWidthCm"])
plt.subplot(2,2,2)
sns.scatterplot(data=data, x=data.index, y=data["PetalWidthCm"], hue="Species")
plt.subplot(2,2,3)
sns.boxplot(data['PetalWidthCm'])
```

```python
X = data.drop("Species", axis=1)
y = data["Species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb.fit(Xtrain, ytrain)

ypred = xgb.predict(Xtest)
```

```python
confusion_matrix(ytest, ypred)
```

```python
print(classification_report(ytest, ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators = 20, random_state=0)
rfc.fit(Xtrain, ytrain)

ypred1 = rfc.predict(Xtest)
```

```python
confusion_matrix(ytest, ypred1)
```

```python
print(classification_report(ytest, ypred1))
```
